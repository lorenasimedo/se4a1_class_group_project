```python
import os
import json

def find_file(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)

def read_file_lines(file_name, path):
    path = find_file(file_name, path)
    if path:
        with open(path, 'r', encoding='utf-8', errors='ignore') as file:
            lines =file.readlines()
        file.close()
        return lines
    return []

def json_file_to_dict(file):
    data = {}
    with open(os.path.join(file), 'r') as fp:
        data = json.load(fp)
    fp.close()
    return data

def extract_project_issues_info(project):
    
    data_file = base_path + "data/" + project + "/bugs_from_github_issues.json"
    issues_non_dvc_file = "output_Non_DVC_related_issues"
    issues_dvc_file  = "output_DVC_related_issues"
    
    lines_non_dvc = read_file_lines(issues_non_dvc_file, output_folder)
    lines_dvc = read_file_lines(issues_dvc_file, output_folder)
    data = json_file_to_dict(data_file)
    
    non_dvc_issues_list = []
    dvc_issues_list = []

    for l in lines_non_dvc:
        if "," + project in l:
            non_dvc_issues_list.append(l.split(",")[0])
            
    
    for l in lines_dvc:
        if "," + project in l:
            dvc_issues_list.append(l.split(",")[0])

    total_hours = 0
    total_comments = 0
    dvc_hours = 0
    dvc_comments = 0
    non_dvc_hours = 0
    non_dvc_comments = 0
    
    for issue in non_dvc_issues_list:
        
        total_hours += float(data[issue]["hours_to_resolution"])
        non_dvc_hours += float(data[issue]["hours_to_resolution"])
        
        total_comments  += int(data[issue]["number_of_comments"])
        non_dvc_comments += int(data[issue]["number_of_comments"])
    
    
    for issue in dvc_issues_list:
        
        total_hours  += float(data[issue]["hours_to_resolution"])
        dvc_hours += float(data[issue]["hours_to_resolution"])
        
        total_comments  += int(data[issue]["number_of_comments"])
        dvc_comments += int(data[issue]["number_of_comments"])
        
    total_hours_average = total_hours / (len(non_dvc_issues_list) + len(dvc_issues_list))
    non_dvc_hours_average = non_dvc_hours / len(non_dvc_issues_list)
    dvc_hours_average = dvc_hours / len(dvc_issues_list)
      
    total_comments_average = total_comments / (len(non_dvc_issues_list) + len(dvc_issues_list))
    non_dvc_comments_average = non_dvc_comments / len(non_dvc_issues_list)
    dvc_comments_average = dvc_comments / len(dvc_issues_list)
    
    print("Hours to resolution total average: " + str(total_hours_average))
    print("Non-dvc hours to resolution average: " + str(non_dvc_hours_average))
    print("DVC hours to resolution average: " + str(dvc_hours_average))
    
    print("----")
     
    print("Comments total average: " + str(total_comments_average))
    print("Non-dvc comments average: " + str(non_dvc_comments_average))
    print("DVC comments average: " + str(dvc_comments_average))
    
extract_project_issues_info("pygmt")
                                 
            
            
    
```
